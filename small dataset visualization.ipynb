```python
# =========================================================
# Employee Dataset Analysis & Visualization (Small Dataset)
# Small Dataset Visualization

#**Author:** Tina  
#**Purpose:** Demonstrate how to visualize a small employee dataset using Python this data is from Kaggle Website

# =========================================================
# This script demonstrates how to visualize a small employee dataset
# using Python's pandas, matplotlib, and numpy libraries.
# It includes scatter plots, bar charts, trend lines, correlation heatmaps, and scatter matrix.
### 1. Load Dataset
## 2. Scatter Plot: Salary vs Productivity
## 3. Bar Chart: Department vs Salary
## 4. Scatter Plot: Experience vs Salary with Trend Line
## 5. Correlation Heatmap
## 6. Scatter Matrix (Pair Plot)

# =========================================================

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ==========================
# 1. Load the Dataset
# ==========================
file_path = "/Users/fatemeeskandarighargheh/Downloads/employee_productivity.csv"
df = pd.read_csv(file_path)
print("âœ… Dataset Loaded Successfully!")
print(df.head())

# ==========================
# 2. Scatter Plot: Salary vs Productivity
# ==========================
plt.figure(figsize=(8, 6))
plt.scatter(df['Productivity_Score'], df['Salary_USD'], color='blue', s=100, edgecolor='black')
plt.title("Salary vs Productivity", fontsize=16)
plt.xlabel("Productivity Score", fontsize=12)
plt.ylabel("Salary (USD)", fontsize=12)

# Annotate Employee IDs
for i, emp_id in enumerate(df['Employee_ID']):
    plt.annotate(emp_id, (df['Productivity_Score'][i], df['Salary_USD'][i] + 500))

plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("salary_vs_productivity.png")
plt.show()

# ==========================
# 3. Bar Chart: Department vs Salary
# ==========================
plt.figure(figsize=(8, 6))
plt.bar(df['Department'], df['Salary_USD'], color='skyblue', edgecolor='black')
plt.title("Department vs Salary", fontsize=16)
plt.xlabel("Department", fontsize=12)
plt.ylabel("Salary (USD)", fontsize=12)
plt.xticks(rotation=30, ha='right')

# Annotate salary values
for i, salary in enumerate(df['Salary_USD']):
    plt.text(i, salary + 1000, f"${salary:,}", ha='center', fontsize=10)

plt.tight_layout()
plt.savefig("department_vs_salary.png")
plt.show()

# ==========================
# 4. Scatter Plot: Experience vs Salary with Trend Line
# ==========================
plt.figure(figsize=(8, 6))
plt.scatter(df['Experience_Years'], df['Salary_USD'], color='orange', s=100, edgecolor='black')

# Annotate Employee IDs
for i, emp_id in enumerate(df['Employee_ID']):
    plt.annotate(emp_id, (df['Experience_Years'][i] + 0.05, df['Salary_USD'][i] + 500))

# Add trend line
x = df['Experience_Years']
y = df['Salary_USD']
coeffs = np.polyfit(x, y, 1)
trend_line = np.poly1d(coeffs)
plt.plot(x, trend_line(x), color='red', linestyle='--', linewidth=2, label='Trend Line')

plt.title("Experience vs Salary", fontsize=16)
plt.xlabel("Experience (Years)", fontsize=12)
plt.ylabel("Salary (USD)", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("experience_vs_salary.png")
plt.show()

# ==========================
# 5. Correlation Heatmap
# ==========================
numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Correlation')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45, ha='right')
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title("Correlation Heatmap of Employee Metrics", fontsize=16)
plt.tight_layout()
plt.savefig("correlation_heatmap.png")
plt.show()

print("Correlation Matrix:")
print(correlation_matrix)

# ==========================
# 6. Scatter Matrix (Pair Plot)
# ==========================
pd.plotting.scatter_matrix(numeric_df, figsize=(14, 14), diagonal='hist', alpha=0.8, color='green')
plt.suptitle("Scatter Matrix of Employee Metrics", y=1.02, fontsize=18)
plt.savefig("scatter_matrix.png")
plt.show()

print("âœ… Analysis complete! All plots have been saved as images.")

## Conclusion

#- Higher productivity generally correlates with higher salary.  
#- Experience also tends to increase salary, as shown by the trend line.  
#- Correlation heatmap highlights which metrics are closely related.  
#- This notebook demonstrates how to analyze and visualize a **small dataset** clearly.
```
